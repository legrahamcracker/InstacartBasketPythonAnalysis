```python
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

## 4.8.1.2 Create Loyalty Flag

```python
# create new column 'max_order' using the max number order number per user ID

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# create loyalty flag based on loc() function on max order per user ID

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge.groupby('loyalty_flag')['prices'].mean()
```

## 4.8.1.2 Create Spender Flag

```python
# create average order price column using mean of product price per user ID

ords_prods_merge['average_order_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(35)
```

```python
# create spender flag using loc() function on average product price per user ID

ords_prods_merge.loc[ords_prods_merge['average_order_price'] >= 10, 'spender_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['average_order_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge['spender_flag'].value_counts(False)
```

## 4.8.1.3 Create Order Frequency Flag

```python
# create 'order_frequency_median' column with median of 'days_since_prior_order' per user ID

ords_prods_merge['order_frequency_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# create 'order_frequency_median' column based on 'order_frequency_median'

ords_prods_merge.loc[ords_prods_merge['order_frequency_median'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_frequency_median'] <= 20) & (ords_prods_merge['order_frequency_median'] > 10), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency_median'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(False)
```
